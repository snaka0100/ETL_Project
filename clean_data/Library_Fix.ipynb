```python
import pandas as pd
import csv
from sqlalchemy import create_engine
```

```python
libraries = pd.read_csv('../Data_Sources/librarymap_by_country.csv', encoding="ISO-8859-1")
libraries.head()
```

```python
libraries_df = libraries[["Library Type", "Country or Area", "Metric", "Metric Data"]]
libraries_df
```

```python
libraries_df = libraries_df.rename(columns={"Country or Area":"Country", "Metric Data":"Number of Libraries"})
libraries_df.head()
```

```python
libraries_only = libraries_df.loc[libraries_df["Metric"]=="Libraries"]
libraries_only.head()
```

```python
#replacing Nan with 0, inplace means that your will be completly replacing the df
libraries_only.fillna(0,inplace = True)
#this is another way to replace the old df with the new one 
#libraries_only = libraries_only.fillna(0,inplace = True)
```

```python
#replacing all '-' with zero
libraries_only.replace('-','0',inplace = True)
```

```python
libraries_only["Number of Libraries"] = libraries_only["Number of Libraries"].astype(float)
libraries_only.dtypes
```

```python
library_group = libraries_only.groupby(libraries_only["Country"])
library_count = pd.DataFrame(library_group["Number of Libraries"].sum())
```
